```python
from models.mixnet.mixnet import MixNet
import torch
import torchvision
import torch.nn as nn
import torch.nn.functional as F
import cv2
import numpy as np
from config import CFG
```

```python
def xyxy2xywh(box):
    x1, y1, x2, y2 = box.T
    new_box = torch.stack((x1, y1, x2 - x1, y2 - y1), dim=1)
    return new_box
```

```python
from dataset.KittiDataset import KittiDataset
from torch.utils.data.dataloader import DataLoader
from torch.utils.data.sampler import SubsetRandomSampler
import matplotlib.pyplot as plt
import matplotlib.patches as patches


dataset = KittiDataset(r"dataset/train.csv", r"dataset/kitti_single/training")
image, target = dataset[456]

boxes = xyxy2xywh(target["boxes"]).detach().numpy()

# Create figure and axes
fig, ax = plt.subplots(1)

fig.set_figheight(20)
fig.set_figwidth(20)
# Display the image
ax.imshow(image.permute(1, 2, 0).detach().numpy())

for box in boxes:
    # Create a Rectangle patch
    rect = patches.Rectangle((box[0],box[1]),box[2],box[3],linewidth=2,edgecolor='r',facecolor='none')
    # Add the patch to the Axes
    ax.add_patch(rect)

plt.show()
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
from models.net import DistanceEstimation
from utils.utils import train_one_epoch
from torch.nn import CrossEntropyLoss, SmoothL1Loss
from torch.optim import AdamW
from torch.optim.lr_scheduler import ExponentialLR
from torch.nn import MSELoss


dataloader_train = DataLoader(dataset, batch_size=CFG["batch_size"], shuffle=True, 
                              num_workers=0, drop_last=True,
                              collate_fn=lambda x: tuple(zip(*x)))

model = DistanceEstimation(img_sizes=450, device=device)
model.to(device)

crossentropy_loss = CrossEntropyLoss()
smooth_loss = SmoothL1Loss()
metric = MSELoss()
optimizer = AdamW(model.parameters(), lr=CFG["lr"])
lr_scheduler = ExponentialLR(optimizer, gamma=0.1, verbose=True)
```

```python
num_train = len(dataset)
indices = list(range(num_train))
np.random.shuffle(indices)

split = int(np.floor(CFG["valid_size"] * num_train))
train_idx, valid_idx = indices[split:], indices[:split]

# define samplers for obtaining training and validation batches
train_sampler = SubsetRandomSampler(train_idx)
valid_sampler = SubsetRandomSampler(valid_idx)

train_loader = DataLoader(dataset, batch_size=CFG["batch_size"], num_workers=0, drop_last=True, 
                          sampler=train_sampler, collate_fn=lambda x: tuple(zip(*x)))

valid_loader = DataLoader(dataset, batch_size=CFG["batch_size"],num_workers=0, drop_last=True, 
                          sampler=valid_sampler, collate_fn=lambda x: tuple(zip(*x)))
```

```python
best_val_loss = np.inf

for epoch in range(CFG["epochs"]):
    val_loss = train_one_epoch(model, optimizer, train_loader, valid_loader,
                              smooth_loss, crossentropy_loss, epoch, device, metric)
    if epoch % 10 == 9:
        lr_scheduler.step()
    if best_val_loss > val_loss:
        best_val_loss = val_loss
        torch.save(model.state_dict(), f"dist_model_val{best_val_loss}.pth")
```
